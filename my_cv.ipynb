```python
#resize

import cv2 as cv

path=r"C:\Users\MMM\Downloads\CV-lab\Kingfisher.jpeg"
img=cv.imread(path)

dim=(100,100)
resized_img=cv.resize(img,dim,interpolation = cv.INTER_AREA)
cv.imshow('Kingfisher',resized_img)
cv.waitKey(0)
```
